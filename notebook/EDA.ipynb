```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/car data.csv")

df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

Price Distribution

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Selling_Price'], kde = True, color = 'blue', bins = 30)
plt.title('Distribution of Selling Price', fontsize = 15)
plt.savefig("../images/Distribution_Of_Selling_Price.png")
plt.show()
```

Correlation HeatMap

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap', fontsize = 15)
plt.savefig("../images/Correlation_Heatmap.png")
plt.show()
```

Box Plot :

```python
sns.boxplot(x='Fuel_Type', y='Selling_Price', data=df)
plt.title('Box Plot of Selling Price by Fuel Type', fontsize = 15)
plt.savefig("../images/Box_Plot.png")
plt.show()
```

Fuel Type Distribution

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x="Fuel_Type",
    palette="Set2"
)

plt.title("Fuel Type Distribution")

plt.tight_layout()

plt.savefig("../images/fuel_type_distribution.png", dpi=300)

plt.show()
```

Selling Type Distruibution

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x="Selling_type",
    palette="pastel"
)

plt.title("Seller Type Distribution")

plt.tight_layout()

plt.savefig("../images/seller_type_distribution.png", dpi=300)

plt.show()
```

Transmission Distribution

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x="Transmission",
    palette="viridis"
)

plt.title("Transmission Distribution")

plt.tight_layout()

plt.savefig("../images/transmission_distribution.png", dpi=300)

plt.show()
```

Present Price vs Selling Price

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="Present_Price",
    y="Selling_Price"
)

plt.title("Present Price vs Selling Price")

plt.tight_layout()

plt.savefig("../images/present_vs_selling_price.png", dpi=300)

plt.show()
```
